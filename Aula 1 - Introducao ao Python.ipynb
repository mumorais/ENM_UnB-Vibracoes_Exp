Neste notebook, vamos documentar algumas funções fundamentais para as aulas práticas de vibrações.

A equação do movimento de um sistema massa mola de 1 Grau de Liberdade, com massa $m$ e rigidez $k$ para descrito por $x$.

$$
m \ddot{x}(t) + k x(t) = 0
$$

Vamos importar a bibliotecas básicas de python para o curso de Vibrações I

```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
```

Vamos plotar uma função senoidal do tipo
$$
x(t) = A \sin (\omega t - \phi)
$$

```python
dt = 1e-3    #passo de tempo, segundos
t0 = 1       #tempo inicial, segundos
tf = 10      #tempo final, segundos
Npts = (tf - t0)/dt + 1   #número de pontos do vetor tempo

#vetor de tempo
tempo = np.arange(0,Npts)*dt + t0
```

Com o vetor tempo, vamos calcular a função senoidal $x(t)$

```python
#parametros
A = 1e-3              #amplitude, metros
freq = 20             #frequência, Hz
omega = 2*np.pi*freq  #frequência, rad/s
T = 1/freq            #período de oscilação, segundos
phase = 0             #fase, rad

x = A*np.sin(omega * tempo - phase)
```

Vamos plotar o gráfico de $x(t)$ em função do tempo $t$

```python
#Faz o gráfico com N periodos
Nperiodos = 2

plt.figure(figsize = [6,2])
plt.plot(tempo, x, label = r'$x(t)$')
plt.xlim((t0, t0 + Nperiodos*T))
plt.xlabel('Tempo, s')
plt.ylabel(r'$x(t)$')
plt.grid()
plt.legend()
```

vamos verficar o efeito da mudança de $\omega$ no comportamento da resposta

```python
frequencias = [20, 100, 500]  #lista de frequencias, rad/s
Nperiodos = 10
plt.figure()
for omega in frequencias:
    x = A*np.sin(omega*tempo - phase)
    plt.plot(tempo, x, label = r'$\omega = $' + str(omega) + ' rad/s')

plt.xlim((t0, t0 + Nperiodos*T))
plt.xlabel('Tempo, s')
plt.ylabel(r'$x(t)$')
plt.grid()
plt.legend()
```

vamos verificar o efeito da mudança da fase $\phi$

```python
fases = [0, np.pi/4, np.pi/2, np.pi, 2*np.pi]  #lista de fases, rad
Nperiodos = 2
omega = 2*np.pi*20 #frequencia, rad/s
plt.figure()
for fase in fases:
    x = A*np.sin(omega*tempo - fase)
    tphi = fase/omega                    #tempo de atraso, segundos
    plt.plot(tempo, x, label = r'$\phi = $ %2.2f $\pi$ rad, $t_{\phi}$ = %1.3f s' % (fase/np.pi, tphi))

plt.xlim((t0, t0 + Nperiodos*T))
plt.xlabel('Tempo, s')
plt.ylabel(r'$x(t)$')
plt.grid()
plt.legend()
```

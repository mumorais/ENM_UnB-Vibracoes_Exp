# 4. Funções
As funções são definidas usando o comando `def`, seguido por um nome de função (por exemplo `my_function_1`). O nome da função é seguido por parênteses, contendo os parâmetros de entrada da função (por exemplo `my_param_1` e `my_param_2`).
No final da definição da função de primeira linha há dois pontos - `:`.

Na linha seguinte, escrevemos nosso código, algoritmos e realizamos cálculos. Todo o código deve ser deslocado uma guia para a direita em relação ao recuo `def`. É assim que o código está associado à função específica.

A saída da função é definida usando o comando `return`, seguido pelas variáveis de saída desejadas.

Para uma melhor compreensão do código é útil fornecer documentação para cada função. Documentação - muitas vezes chamada de "docstring" é definida no seguinte formato `""" Documentação """" `.

````python
def my_function_1(my_param_1, my_param_2=2):
    """
    Returns sum of my_param_1 and my_param_2.

    :parm my_param_1: the first parameter of summation
    :parm my_param_2: the second parameter has a defined default value, meaning that if this parameter
                      is not defined, the function will use the default value (my_param_2=2) for execution.

    :return: sum of my_param_1 + my_param_2

    Example:
    >>>a = my_function_1(1, 1)
    >>>print(a)
    2
    """
    result = my_param_1 + my_param_2
    return result
````

# 5. Classes/Objects
Python é uma linguagem de programação orientada a objetos. Quase tudo em Python é um objeto, com suas propriedades e métodos.

````python
class Participant:
    """
    Here is provided docstrings for class.
    """

    def __init__(self, name, surname):
        """
        Here docstrings are provided for the `__init__` method. The input parameters must also be described.

        :parm name: name of the person
        :parm surname: surname of the person

        """
        self.name = name
        self.surname = surname

    def get_full_name(self):
        """
        Here docstrings are provided for the `get_full_name` method.

        :return: full name of the participant - name and surname combined
        """
        self.full_name = self.name + " " + self.surname
        return self.full_name

    def say_welcome(self):
        """
        Say welcome to the participant

        :return: welcome to the participant
        """
        self.welcome = "Welcome " + self.name + "!"
        return self.welcome

    def say_welcome_full(self):
        """
        Say welcome to the participant

        :return: welcome to the participant
        """
        self.welcome_full = "Welcome " + self.name +" "+ self.surname + "!"
        return self.welcome_full
````

````python
first_participant = Participant('Isac','Newton')
````

````python
first_participant.get_full_name()
````

````python
first_participant.say_welcome()
````

````python
first_participant.say_welcome_full()
````

## 6.3  Basic operation with numpy arrays

````python
A = np.array([[1, 1],
              [0, 1]])
B = np.array([[2, 0],
              [3, 4]])

A + B # elementwise summation
A - B # elementwise subtraction
A * B # elementwise product
A @ B  # matrix product
np.linalg.inv(A) # inverse
np.linalg.pinv(A) # pseudo-inverse
np.transpose(B) # transpose
````

## 7.3 While loop

````python
# example of while loop
i = 0
while i < 6:
    print(i)
    if i == 3:
        break
    i += 2 # this is shorter form for i = i + 1

print("i = ", i)
````

# 8. Simple ploting with matplotlib

Para plotagem geralmente usamos a biblioteca [matplotlib](https://matplotlib.org/). É simples de usar, mas ainda permite a criação de gráficos avançados e bonitos.

````python
import matplotlib.pyplot as plt # this is also a convention, for basic ploting is enough if we only import the ``pyplot`` part

# basic usage - closer to matlab form
x = np.linspace(0, 6*np.pi, 1000)
y = np.sin(x) + np.cos(2*x)
plt.plot(x, y)
plt.xlim(5, 15)
plt.ylim(-2, 2)
plt.grid()
plt.xlabel("$x$ [s]")
plt.ylabel("$y$ [m]")
````

````python
# more pythonic way to plot simple or advanced plots, by creating objects
fig, ax = plt.subplots(1, 1, figsize = (3, 3)) # this command gives you plenty of advanced possibilities of ploting
ax.plot(x, y)
ax.set_xlim(5, 15)
ax.set_ylim(-2, 2)
ax.grid()
ax.set_xlabel("$x$ [s]")
ax.set_ylabel("$y$ [m]")
````
